```python
from forces import Forces3D
import numpy as np
import pickle
import matplotlib.pyplot as plt
import utility_forces as uf
from marching_cubes import Mesh3D
```

## Curva

```python
with open('rusienkiewicz.pkl', 'rb') as archivo:
    rusinki = pickle.load(archivo)
```

```python
valores_calculados = rusinki['calc_curvatures'][::-1]
valores_esperados = rusinki['real_curvatures'][::-1]
radios = rusinki['teoric_radius'][::-1]

mse = np.sqrt(np.mean((valores_esperados - valores_calculados) ** 2))

# 3. Gr√°fico comparativo con banda de error
plt.figure(figsize=(10, 6))
plt.plot(radios, valores_esperados, '.', label='Valores Esperados', color='blue')
plt.plot(radios, valores_calculados, '.',  label='Valores Calculados (SCIAN-Forces)', color='red')

plt.title(f'Comparativa de Valores Esperados vs Calculados')
plt.xlabel('radius')
plt.ylabel('curvatures')
plt.legend()
plt.grid(True)

# Mostrar la gr√°fica
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(radios, np.abs(valores_esperados - valores_calculados), color='green')
plt.title(f'Error with MSE={mse:.4f}')
plt.xlabel('radius')
plt.ylabel('error')
plt.grid(True)
plt.show()
```
